```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv(".\List_company_23052023 - Listing.csv")
df1 = pd.read_csv(".\List_company_23052023 - Delisting_after_2014.csv")
df2 = pd.read_csv(".\List_company_23052023 - Delisting_before_2014.csv")

lst_ = list(df['Symbol']) + list(df1['Symbol']) + list(df2['Symbol'])
len(lst_)
```

```python
fields = ['Symbol',	'table1',	'table2',	'table3']

df_ = []
for comp in tqdm(lst_):
    compinfo = pd.read_csv(f"./CompanyInfo/{comp}.csv")
    keys = compinfo['Type'].values
    temp ={}

    temp[fields[0]] = comp
    if '市場' in keys:
        temp[fields[1]] = 1
    else: 
        temp[fields[1]] = 0

    if '決算' in keys:
        temp[fields[2]] = 1
    else: 
        temp[fields[2]] = 0

    if 'EBITDA･会予' in keys:
        temp[fields[3]] = 1
    else: 
        temp[fields[3]] = 0
    
    df_.append(temp)
```

```python
df_ = pd.DataFrame(df_)
```

```python
df_.to_csv(f"./Checklist.csv", header = True ,index = False)
```
