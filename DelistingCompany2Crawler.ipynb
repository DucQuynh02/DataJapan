```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
import pandas as pd
from time import sleep
```

```python
br = webdriver.Edge()

fields = ['上場廃止日',	'銘柄 コード',	'銘柄名',	'理由',	'売買最終日']

df = []

br.get('https://kabu.com/investment/meigara/jyoujyou_haisi.html')

sleep(1)
tb = br.find_element(By.XPATH, './/*[@id="main"]/table')
sleep(1)

rows = tb.find_elements(By.XPATH, ".//tbody/tr")
for row in range(1,len(rows)+1):
        temp = {}
        for col in range(1,6):
            value = tb.find_element(By.XPATH, f".//tbody/tr[{row}]/td[{col}]").text
            temp[fields[col-1]] = value

        df.append(temp)

br.close()
```

```python
pd.DataFrame(df)
```

```python
filename = "Delisting2Company.csv"

with open(filename, 'w', encoding= 'utf-8', newline= '') as csvfile:
    writer = csv.DictWriter(csvfile, fields,lineterminator = '\r')
    writer.writeheader()
    writer.writerows(df)
```
