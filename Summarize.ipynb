```python
import pandas as pd
from tqdm import tqdm
```

```python
fields = ['Symbol',	'Listing',	'Delisting', 'Sàn', 'VolumeNow', 'VolumeListing', 'Ngành']

df_ = []

```

```python


df = pd.read_csv(".\List_company_23052023 - Listing.csv")
lst_ = df['Symbol']
for comp in tqdm(lst_):
    compinfo = pd.read_csv(f"./CompanyInfo/{comp}.csv").set_index('Type')

    keys = compinfo.index
    temp ={}

    temp[fields[0]] = comp
    if '上場日' in keys:
        temp[fields[1]] = compinfo['Value']['上場日']
    

    temp[fields[2]] = '2023/10/10'

    if '市場' in keys:
        temp[fields[3]] = compinfo['Value']['市場']

    if '株数(自己株控除後)' in keys:
        temp[fields[4]] = compinfo['Value']['株数(自己株控除後)']
    
    if '発行済株数' in keys:
        temp[fields[5]] = compinfo['Value']['発行済株数'] 

    if '業種' in keys:
        temp[fields[6]] = compinfo['Value']['業種']
    
    df_.append(temp)
```

```python
df = pd.read_csv(".\List_company_23052023 - Delisting_after_2014.csv").set_index('Symbol')
lst_ = df.index

df
```

```python

for comp in tqdm(lst_):
    compinfo = pd.read_csv(f"./CompanyInfo/{comp}.csv").set_index('Type')

    keys = compinfo.index
    temp ={}

    temp[fields[0]] = comp
    if '上場日' in keys:
        temp[fields[1]] = compinfo['Value']['上場日']
    

    temp[fields[2]] = df['上場廃止日'][comp]
    
    temp[fields[3]] = df['市場区分'][comp]



    if '株数(自己株控除後)' in keys:
        temp[fields[4]] = compinfo['Value']['株数(自己株控除後)']

    if '発行済株数' in keys:
        temp[fields[5]] = compinfo['Value']['発行済株数'] 

    if '業種' in keys:
        temp[fields[6]] = compinfo['Value']['業種']
    
    df_.append(temp)
```

```python
df = pd.read_csv(".\List_company_23052023 - Delisting_before_2014.csv").set_index('Symbol')
lst_ = df.index

df
```
