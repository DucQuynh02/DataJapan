```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv
import pandas as pd
from time import sleep
import time
```

```python
br = webdriver.Edge()

fields = ['Symbol',	'Name',	'Price (Intraday)',
          	'Change',	'% Change',	'Volume',
            'Avg Vol (3 month)',  'Market Cap',
            'PE Ratio (TTM)',	'52 Week Range']

df = []
offset = 0
while True:
    try:
        br.get(f'https://finance.yahoo.com/screener/unsaved/344d9065-e9e0-4f4a-8284-f97aeb6506d2?offset={offset}&count=100')
    except:
         break
    sleep(1)
    
    try:
        tb = br.find_element(By.XPATH ,'.//*[@id="scr-res-table"]/div[1]/table')
    except:
        break
    sleep(1)
    
    rows = tb.find_elements(By.XPATH, ".//tbody/tr")

    for row in range(1,len(rows)+1):
            temp = {}
            for col in range(1,11):
                value = tb.find_element(By.XPATH, f".//tbody/tr[{row}]/td[{col}]").text
                temp[fields[col-1]] = value

            df.append(temp)
    offset += 100
br.close()
```
